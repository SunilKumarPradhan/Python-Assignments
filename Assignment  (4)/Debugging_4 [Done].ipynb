```python
def summation(n):
    '''
    Objective: To find the sum of first n positive integers
    Input Parameter: n - numeric value
    Return Value: total - numeric value
    '''
    total = 0
    for count in range(1, n + 1):  # Fix the loop range
        total += count
    return total

def main():
    '''
    Objective: To find sum of first n positive integers based on user input
    Input Parameter: None
    Return Value: None
    '''
    n = int(input("Enter number of terms: "))
    total = summation(n)
    print("Sum of first", n, "positive integers:", total)

if __name__ == '__main__':
    main()
```

```python
def invertedRightTriangle(nRows):
    '''
    Objective: To print an inverted right triangle
    Input Parameter: nRows - integer value
    Return Value: None
    '''
    for i in range(nRows, 0, -1):  # Fix the loop range
        print("*" * i)

def main():
    '''
    Objective: To print an inverted right triangle based on user input
    Input Parameter: None
    Return Value: None
    '''
    nRows = int(input("Enter number of rows: "))
    invertedRightTriangle(nRows)

if __name__ == '__main__':
    main()
```

```python
def main():
    '''
    Objective: To display the percentage of marks scored by the student
    Input Parameter: None
    Return Value: None
    '''
    totalMarks = 0
    i = 0

    while True:
        marks = input("Enter marks for subject " + str(i + 1) + " (press Enter to finish): ")

        if marks == '':
            break

        marks = int(marks)

        if marks < 0 or marks > 100:
            print("Invalid marks")
            continue

        i += 1
        totalMarks += marks

    if i == 0:
        print("No marks entered.")
    else:
        percentage = (totalMarks / i)
        print("Total marks:", int(totalMarks))
        print("Percentage:", round(percentage, 2))

if __name__ == '__main__':
    main()
```

```python
def isLeapYear(year):
    '''
    Objective: To determine whether a given year is a leap year or not
    Input Parameter: year - numeric value
    Return value: True if year is a leap year, False otherwise
    '''
    return (year % 400 == 0) or ((year % 100 != 0) and (year % 4 == 0))

# Testing the function
year_to_check = 2100  # You can change this value for testing different years
result = isLeapYear(year_to_check)
print(f"{year_to_check} is a leap year: {result}")
```

```python
def findHCF(num1, num2):
    '''
    Objective: To find HCF of two numbers, num1 and num2
    Input Parameters: num1, num2 - numeric values
    Return Value: HCF - numeric value
    '''
    if num1 < num2:
        minNum = num1
    else:
        minNum = num2

    for i in range(minNum, 0, -1):  # Include 1 in the range
        if num1 % i == 0 and num2 % i == 0:
            return i

def main():
    '''
    Objective: To take two numbers as input and find their HCF
    Input Parameter: None
    Return Value: None
    '''
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    print(findHCF(num1, num2))

if __name__ == '__main__':
    main()
```
